```python
import papermill as pm
import os   
template_nb = "area2_auto_integrate.ipynb"  
os.makedirs("outputs", exist_ok=True)

monkeys = [
    "Han_20171207",
    "Duncan_20190710",
    
    "Chips_20170913",
    "Han_20171204",
    "Duncan_20191106",
    "Lando_20170803",

    ## These ones are disregarded for now
    # "Lando_20170731",
    # "Han_20171201",
    # "Duncan_20191016",
]

for monkey in monkeys:
    out_nb = f"outputs/{monkey}_actpas_psth_decode.ipynb"
    print("Hello from:", monkey, "->", out_nb)

    try:
        pm.execute_notebook(
            template_nb,
            out_nb,
            parameters={"monkey": monkey},
            kernel_name="sdt_env",     
        )
        print("Finished:", out_nb)
    except Exception as e:
        msg = str(e)
        # Detect intentional stop message
        if "Intentional stop (Papermill)" in msg:
            print("Stopped early on purpose for:", monkey)
            continue
        print("FAILED:", monkey,e)
        continue
```
